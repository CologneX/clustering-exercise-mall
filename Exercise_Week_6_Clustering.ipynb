## Load Dataset

The dataset that we will use is a customer transaction in a certain mall. In this dataset there will 4 features including Gender, Age, Annual Income, and Spending Score. You can get the dataset from https://raw.githubusercontent.com/catharinamega/Data-Mining-ISB-2020/main/Mall_Customers.csv

```python
# load dataset and show the first five sample
df = pd.read_csv('https://raw.githubusercontent.com/catharinamega/Data-Mining-ISB-2020/main/Mall_Customers.csv')
df.head()
```

```python
# describe data
df.describe()
```

## Data Preprocessing

For this case study, we only use income and spending score 

```python
# create new dataframe that consist only annual income and spending score
df_ai_sc = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
df_ai_sc.head()
```

Before we analyze our data, we need to scale the data with standardization method

```python
# scale dataset with standard scaler
scaler = StandardScaler()
df_ai_sc_scaled = scaler.fit_transform(df_ai_sc)
```

```python
# get stastical information on the new dataset
df_ai_sc_scaled = pd.DataFrame(df_ai_sc_scaled, columns=['Annual Income (k$)', 'Spending Score (1-100)'])
df_ai_sc_scaled.describe()
```

Then, we check the inisial position for our sample. You may use pairplot or scattered chart for the visualization option.

```python
# Visualize the position 
plt.scatter(df_ai_sc_scaled['Annual Income (k$)'], df_ai_sc_scaled['Spending Score (1-100)'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
# Visualize the position using the original scale
plt.scatter(df_ai_sc['Annual Income (k$)'], df_ai_sc['Spending Score (1-100)'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()  
```

## Clustering

### K-Means Algorithm

For using KMeans clustering algorithm, you need to define the number of cluster, initialization number, number of iteration, and random state.

In this case study, the below parameter settings is used:
- centroid initialization: random
- number of cluster: 5
- number of iteration: 300
- random state: 42

```python
# assign cluster with K-Means algorithm
kmeans = KMeans(n_clusters=5, random_state=42, max_iter=300, init='random', n_init='auto')
kmeans.fit_predict(df_ai_sc_scaled)
```

Visualize your result based on income and spending. You also can show the final position of the centroid.

```python
# visualize the result using scater chart
plt.scatter(df_ai_sc_scaled['Annual Income (k$)'], df_ai_sc_scaled['Spending Score (1-100)'], c=kmeans.labels_)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

The evaluation for chosing the number of K in KMEANS can be done by several methods such as elbow method and silhouette score

1. using elbow method

```python
# set the parameter for evaluating K
inertia = []
# check the inertia for each K
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, max_iter=300, init='random', n_init='auto')
    kmeans.fit(df_ai_sc_scaled)
    inertia.append(kmeans.inertia_)
```

```python
# manually visualize the elbow method using plotlib
plt.plot(range(1, 11), inertia)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# using kneelocator to find the elbow/knee
kn = KneeLocator(range(1, 11), inertia, curve='convex', direction='decreasing')
print(f"The best k for your model is {kn.knee}")
```
